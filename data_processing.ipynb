```python
ids_number_in_folder_01 = "/workspace/datasets/utteranceEMOTIOn/data-processed-icassp-20/IEMOCAP/seven_category_120/folds/fold01/fold01_id.txt"
```

```python
with open(ids_number_in_folder_01, 'r') as f:
    lines = f.readlines()

lines = [line.strip() for line in lines]
train_ids = lines[0].split(" ")
train_ids = [int(id) for id in train_ids]
dev_ids = lines[1].split(" ")
dev_ids = [int(id) for id in dev_ids]
test_ids = lines[2].split(" ")
test_ids = [int(id) for id in test_ids]

ids = {}

ids['train'] = train_ids
ids['dev'] = dev_ids
ids['test'] = test_ids

modes = ['train', 'dev', 'test']
```

```python
import numpy as np
import os
kkkk=0

# while True:
for mode in modes:
    mfcc_path = f"/workspace/datasets/utteranceEMOTIOn/data-processed-icassp-20/IEMOCAP/seven_category_120/seven_C_MFCC120.npy"
    video_path = f"/workspace/datasets/utteranceEMOTIOn/data-processed-icassp-20/IEMOCAP/seven_category_120/seven_C_video.npy"
    text_path = f"/workspace/datasets/utteranceEMOTIOn/data-processed-icassp-20/IEMOCAP/seven_category_120/folds/fold01/{mode}_text.npy"
    label_path = f"/workspace/datasets/utteranceEMOTIOn/data-processed-icassp-20/IEMOCAP/seven_category_120/folds/fold01/{mode}_label.npy"
    trans_path = f"/workspace/datasets/utteranceEMOTIOn/data-processed-icassp-20/IEMOCAP/seven_category_120/seven_C_trans.txt"
    mfcc = np.load(mfcc_path)
    video = np.load(video_path)
    text = np.load(text_path)
    label = np.load(label_path)
    with open(trans_path, 'r') as f:
        trans = f.readlines()
    
    trans = [line.strip() for line in trans]        
    

    for i in range(len(label)):
        mfcc_feature = mfcc[ids[mode][i]]
        video_feature = video[ids[mode][i]]
        
        text_feature = text[i]
        label_feature = label[i]
        
        trans_feature = trans[ids[mode][i]]
        session_name = str_ids[ids[mode][i]]  # 예: Ses01F_impro01_000
        print(i)
        print(ids[mode][i])
        print(session_name)
        # print(text_feature)
        print(label_feature)
        # print(trans_feature)
        samples_path = os.path.join("/workspace/datasets/MKD_dataset/IEMOCAP_baseline_iscross", session_name)

        # 저장 경로들
        mfcc_save_path = os.path.join(samples_path, "audio", "mfcc_from_iscross", "sample.npy")
        resnet_save_path = os.path.join(samples_path, "video", "resnet_from_iscross", "sample.npy")
        glove_save_path = os.path.join(samples_path, "text", "glove_from_iscross", "sample.npy")
        label_save_path = os.path.join(samples_path, "label", "sample.txt")
        trans_save_path = os.path.join(samples_path, "text", "trans", "sample.txt")

        # 만약 모든 파일이 이미 존재하면 pass
        if (
            os.path.exists(mfcc_save_path) and
            os.path.exists(resnet_save_path) and
            os.path.exists(glove_save_path) and
            os.path.exists(label_save_path) and
            os.path.exists(trans_save_path)
        ):
            kkkk +=1
            print(kkkk)
            print(f"⏩ Already exists, skipped: {session_name}")
            continue

        # 디렉토리 생성
        for path in [mfcc_save_path, resnet_save_path, glove_save_path, label_save_path,trans_save_path]:
            os.makedirs(os.path.dirname(path), exist_ok=True)

        
        
        # 저장
        np.save(mfcc_save_path, mfcc_feature)
        np.save(resnet_save_path, video_feature)
        np.save(glove_save_path, text_feature)

        with open(label_save_path, 'w') as f:
            f.write(str(label_feature))
            
        with open(trans_save_path, "w") as f:
            f.write(trans_feature)

        print(f"✅ Saved new: {session_name} \n")
# import time
# time.sleep(100)
```
